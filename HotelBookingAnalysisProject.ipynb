# Data Analysis and Visualizations

```python
cancelled_count = df['is_canceled'].value_counts()
print(cancelled_count)
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,6))
bar1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels,_ = bar1. get_legend_handles_labels()
#bar1.legend(bbox_to_anchor(1,1))
plt.title('Reservation status per hotel',size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')

```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel['is_repeated_guest'].value_counts()
```

```python
resort_hotel['adr'].mean()
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (16,8))
plt.title('Average daily rate in City Hotel and Resort Hotel', fontsize=18)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize=12)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize = (12,6))
bar1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')
#legend_label,_ = bar1. get_legend_handles_labels()
#bar1.legend(bbox_to_anchor(1,1))
plt.title('Reservation status per month',size=20)
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.legend(['Not Calnceled', 'Canceled'])
plt.show()
```

```python
cancelled_data = df[df['is_canceled']==1]
top_10_countries = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,6))
plt.pie(top_10_countries, labels=top_10_countries.index, autopct='%.2f')
plt.title("Top 10 countries with reservation cancelled")
plt.show()

```

```python
cancelled_data_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_data_adr.reset_index(inplace=True)
cancelled_data_adr.sort_values('reservation_status_date', inplace=True)
```

```python
not_cancelled_data = df[df['is_canceled']==0]

not_cancelled_data_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_data_adr.reset_index(inplace=True)
not_cancelled_data_adr.sort_values('reservation_status_date', inplace=True)
```

```python
plt.figure(figsize = (18,6))
plt.title("Average Daily Rate")
plt.plot(not_cancelled_data_adr['reservation_status_date'], not_cancelled_data_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_data_adr['reservation_status_date'], cancelled_data_adr['adr'], label='Not Cancelled')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```
